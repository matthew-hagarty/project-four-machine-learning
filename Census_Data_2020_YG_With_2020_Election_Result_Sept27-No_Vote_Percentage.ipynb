```python
##filter for only 2=2020
years = [2]

df_cc_2020_cleaned_yr = df_cc_2020.loc[((df_cc_2020['YEAR'] == 2))]
df_cc_2020_cleaned_yr.head()
```

```python
#filter for age above which is AGEGRP 4 through 18, i,e, removed AGEGRP 0 to 3
df_cc_2020_cleaned_yr_age= df_cc_2020_cleaned_yr.loc[((df_cc_2020_cleaned_yr['AGEGRP'] != 0) & (df_cc_2020_cleaned_yr['AGEGRP'] != 1) & \
                                       (df_cc_2020_cleaned_yr['AGEGRP'] != 2) & (df_cc_2020_cleaned_yr['AGEGRP'] != 3))]
df_cc_2020_cleaned_yr_age.head()
```

```python
##Need to Convert all the non-numeric back to numeric

# List of columns to exclude from conversion
exclude_columns = ["STNAME", "CTYNAME"]

# Iterate through the DataFrame's columns
for column in df_cc_2020_cleaned_yr_age.columns:
    if column not in exclude_columns:
        # Use pd.to_numeric to convert the column to int64
        df_cc_2020_cleaned_yr_age[column] = pd.to_numeric(df_cc_2020_cleaned_yr_age[column], errors='coerce').astype('Int64')
```

```python
df_cc_2020_cleaned_yr_age.dtypes
```

```python
df_cc_2020_cleaned_yr_age.head()
```

```python
# Define a dictionary for renaming
year_mapping = {2: 2020}

# Use the 'replace' method to rename values in the 'YEAR' column
df_cc_2020_cleaned_yr_age['YEAR'] = df_cc_2020_cleaned_yr_age['YEAR'].replace(year_mapping)
```

```python
df_cc_2020_cleaned_yr_age.head()
```

```python
df_cc_2020_cleaned_yr_age.columns
```

```python
# Define the age bins
age_bins = [4, 10, float('inf')]  # The last bin represents 50+ age group

# Define labels for the age groups
age_labels = ['15-49', '50+']

# Use pd.cut to create age groups
df_cc_2020_cleaned_yr_age['AGE_GROUP'] = pd.cut(df_cc_2020_cleaned_yr_age['AGEGRP'], bins=age_bins, labels=age_labels, right=False)

# Now, df contains a new column 'AGE_GROUP' with the age group labels
#drop the old AGEGRP 
df_cc_2020_cleaned_yr_age.drop(columns=['AGEGRP'], inplace=True)
```

# Combined cleaned 2000 to 2020 Census data

```python
#combined 2000 to 2020 election data together
# Specify the folder path where CSV files are located
folder_path_2000_2020 = '/Users/yegu/Documents/Project_4_Machine_learning_YG/Cleaned_CC_CSV/Sept27_CSV'


# Get a list of CSV files in the folder
csv_files_2000_2020 = [f for f in os.listdir(folder_path_2000_2020) if f.endswith('.csv')]

# Create an empty DataFrame to store the combined data
combined_2000_2020_df = pd.DataFrame()

# Iterate through the list of CSV files and combine them
for csv_file in csv_files_2000_2020:
    file_path = os.path.join(folder_path_2000_2020, csv_file)
    df = pd.read_csv(file_path, encoding='ISO-8859-1')
    combined_2000_2020_df = combined_2000_2020_df.append(df, ignore_index=True)
print(f'Combined 2000 to 2020 data complete')   
```

```python
#result the index 
combined_2000_2020_df.reset_index(drop=True).head()
```

```python
#drop the COUNTY column 
combined_2000_2020_df.drop(columns=['COUNTY'], inplace=True)
```

```python
#groupby on state, year, stname, age_group
combined_2000_2020_df_gby = combined_2000_2020_df.groupby(['STATE', 'YEAR', 'STNAME', 'AGE_GROUP']).sum().reset_index()
```

```python
combined_2000_2020_df_gby
```

```python
combined_2000_2020_df_gby.columns
```

```python
# Pivot the DataFrame
pivoted_combined_2000_2020_df_gby_agegrp = combined_2000_2020_df_gby.pivot(index=['STATE', 'YEAR', 'STNAME'], columns='AGE_GROUP', values=['TOT_MALE', 'TOT_FEMALE', 'WA_MALE', 'WA_FEMALE', 'BA_MALE', 'BA_FEMALE', 'IA_MALE', 'IA_FEMALE', 'AA_MALE', 'AA_FEMALE', 'NA_MALE', 'NA_FEMALE', 'TOM_MALE', 'TOM_FEMALE', 'NH_MALE', 'NH_FEMALE', 'H_MALE', 'H_FEMALE'])

# Reset the index to make 'STNAME', 'STATE', 'YEAR' regular columns 
pivoted_combined_2000_2020_df_gby_agegrp.reset_index(inplace=True)

# Flatten the MultiIndex columns
pivoted_combined_2000_2020_df_gby_agegrp.columns = [' '.join(col).strip() for col in pivoted_combined_2000_2020_df_gby_agegrp.columns.values]
```

```python
# Group by the specified columns and calculate the sum of 'candidate_vote' and 'total_vote'
df_election_result_gby = df_election_result.groupby(['year', 'state', 'party'])[['candidatevotes', 'totalvotes']].sum().reset_index()
```

```python
df_election_result_gby.reset_index(inplace=True)
```

```python
df_election_result_gby
```

```python
# Filter the grouped DataFrame to include only 'DEMOCRAT' and 'REPUBLICAN' parties
two_party = (df_election_result_gby['party'] == 'DEMOCRAT') | (df_election_result_gby['party'] == 'REPUBLICAN')
df_election_result_two_party = df_election_result_gby[two_party]
```

```python
#print the df
df_election_result_two_party
```

```python
#df_election_result_two_party.loc[df_election_result_two_party['year']==2000]
```

```python
#calculate the winner based on 'candidatevotes', for year, state, encode winner =1
df_election_result_two_party['Winner'] = (
    df_election_result_two_party.groupby(['year', 'state'])['candidatevotes']
    .transform(lambda x: x.eq(x.max()).astype(int))
)
```

```python
df_election_result_two_party
```

```python
df_election_result_two_party['party_won'] = df_election_result_two_party['party'].where(df_election_result_two_party['Winner'] == 1)
```

```python
df_election_result_two_party.head()
```

```python
#fill in the party_won column where value is NaN with the non-NaN value listed for the year and state
df_election_result_two_party['party_won'] = df_election_result_two_party.groupby(['year', 'state'])['party_won'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```
