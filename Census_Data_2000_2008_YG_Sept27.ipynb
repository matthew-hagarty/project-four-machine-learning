# Read 2000_2008 CSV and Combine the CSVs

```python
#Merge the 2008 census data for all states into one csv 
# Specify the folder path where CSV files are located
folder_path_2000_2008 = '/Users/yegu/Documents/Project_4_Machine_learning_YG/Resources/Census/2008'


# Get a list of CSV files in the folder
csv_files_2000_2008 = [f for f in os.listdir(folder_path_2000_2008) if f.endswith('.csv')]

# Create an empty DataFrame to store the combined data
merged_2008_df = pd.DataFrame()

# Iterate through the list of CSV files and combine them
for csv_file in csv_files_2000_2008:
    file_path = os.path.join(folder_path_2000_2008, csv_file)
    df = pd.read_csv(file_path, encoding='ISO-8859-1')
    merged_2008_df = merged_2008_df.append(df, ignore_index=True)
print(f'Combined data complete')    

# Write the merged data to a new CSV file
#output_file = '/Users/yegu/Documents/Project_4_Machine_learning_YG/Resources/Census/2008_merged_census_data.csv'
#merged_2008_df.to_csv(output_file, index=False, encoding='utf-8')
#print(f'Merged data saved to {output_file}')
```

```python
## Key to the numbers for census data: https://www2.census.gov/programs-surveys/popest/technical-documentation/file-layouts/2000-2008/cc-est2008-alldata.pdf
##filter for only 3=2000, 7=2004,11=2008
years = [3, 7, 11]

df_2000_2008_cleaned_yr = df_2000_2008_cleaned.loc[((df_2000_2008_cleaned['YEAR'] == 3) | (df_2000_2008_cleaned['YEAR'] == 7) | \
                                       (df_2000_2008_cleaned['YEAR'] == 11))]
df_2000_2008_cleaned_yr.head()
```

```python
#filter for age above which is AGEGRP 4 through 18, i,e, removed AGEGRP 0 to 3
df_2008_cleaned_yr_age = df_2000_2008_cleaned_yr.loc[((df_2000_2008_cleaned_yr['AGEGRP'] != 0) & (df_2000_2008_cleaned_yr['AGEGRP'] != 1) & \
                                       (df_2000_2008_cleaned_yr['AGEGRP'] != 2) & (df_2000_2008_cleaned_yr['AGEGRP'] != 3))]
df_2008_cleaned_yr_age
```

```python
##Need to Convert all the non-numeric back to numeric

# List of columns to exclude from conversion
exclude_columns = ["STNAME", "CTYNAME"]

# Iterate through the DataFrame's columns
for column in df_2008_cleaned_yr_age.columns:
    if column not in exclude_columns:
        # Use pd.to_numeric to convert the column to int64
        df_2008_cleaned_yr_age[column] = pd.to_numeric(df_2008_cleaned_yr_age[column], errors='coerce').astype('Int64')
```

```python
# Define the age bins
age_bins = [4, 10, float('inf')]  # The last bin represents 50+ age group

# Define labels for the age groups
age_labels = ['15-49', '50+']

# Use pd.cut to create age groups
df_2008_cleaned_yr_age['AGE_GROUP'] = pd.cut(df_2008_cleaned_yr_age['AGEGRP'], bins=age_bins, labels=age_labels, right=False)

# Now, df contains a new column 'AGE_GROUP' with the age group labels
#drop the old AGEGRP 
df_2008_cleaned_yr_age.drop(columns=['AGEGRP'], inplace=True)
```
