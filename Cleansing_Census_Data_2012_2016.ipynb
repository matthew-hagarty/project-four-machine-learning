# Read 2012_2016 Census Data

```python
#keys to the cc data https://www2.census.gov/programs-surveys/popest/technical-documentation/file-layouts/2010-2016/cc-est2016-alldata.pdf
```

```python
folder_path_2016 = '/Users/yegu/Documents/Project_4_Machine_learning_YG/Resources/Census/cc-est2016-alldata.csv'
df_cc_2012_2016 = pd.read_csv(folder_path_2016, encoding='ISO-8859-1')
df_cc_2012_2016.head()
```

```python
distinct_values = df_cc_2012_2016['YEAR'].unique()
print(distinct_values)
#5=2012, 9=2016
```

```python
# Keep: State, County, tot_pop, 'WA_MALE', 'WA_FEMALE', 'BA_MALE',
##       'BA_FEMALE', 'IA_MALE', 'IA_FEMALE', 'AA_MALE', 'AA_FEMALE', 'NA_MALE',
##       'NA_FEMALE', 'TOM_MALE', 'TOM_FEMALE', 'NH_MALE', 'NH_FEMALE', 'H_MALE', 'H_FEMALE',
## Keep Year: keep 3, 7, 11
## Keep AgeGrp: Drop 0-3
df_cc_2012_2016 = df_cc_2012_2016[['STATE','COUNTY','STNAME', 'CTYNAME', 'YEAR', 'AGEGRP', 'TOT_MALE','TOT_FEMALE','WA_MALE', 'WA_FEMALE', 'BA_MALE',
       'BA_FEMALE', 'IA_MALE', 'IA_FEMALE', 'AA_MALE', 'AA_FEMALE', 'NA_MALE',
       'NA_FEMALE', 'TOM_MALE', 'TOM_FEMALE', 'NH_MALE', 'NH_FEMALE', 'H_MALE', 'H_FEMALE']]

df_cc_2012_2016.head()

## What would happen if we kept/excluded the younger age groups
```

```python
##filter for only 5=2012, 9=2016
years = [5,9]

df_cc_2012_2016_cleaned_yr = df_cc_2012_2016.loc[((df_cc_2012_2016['YEAR'] == 5) | (df_cc_2012_2016['YEAR'] == 9))]
df_cc_2012_2016_cleaned_yr.head()
```

```python
#filter for age above which is AGEGRP 4 through 18, i,e, removed AGEGRP 0 to 3
df_cc_2012_2016_cleaned_yr_age= df_cc_2012_2016_cleaned_yr.loc[((df_cc_2012_2016_cleaned_yr['AGEGRP'] != 0) & (df_cc_2012_2016_cleaned_yr['AGEGRP'] != 1) & \
                                       (df_cc_2012_2016_cleaned_yr['AGEGRP'] != 2) & (df_cc_2012_2016_cleaned_yr['AGEGRP'] != 3))]
df_cc_2012_2016_cleaned_yr_age.head()
```

```python
##Need to Convert all the non-numeric back to numeric

# List of columns to exclude from conversion
exclude_columns = ["STNAME", "CTYNAME"]

# Iterate through the DataFrame's columns
for column in df_cc_2012_2016_cleaned_yr_age.columns:
    if column not in exclude_columns:
        # Use pd.to_numeric to convert the column to int64
        df_cc_2012_2016_cleaned_yr_age[column] = pd.to_numeric(df_cc_2012_2016_cleaned_yr_age[column], errors='coerce').astype('Int64')
```

```python
df_cc_2012_2016_cleaned_yr_age.dtypes
```

```python
df_cc_2012_2016_cleaned_yr_age.head()
```

```python
# Define a dictionary for renaming
year_mapping = {5: 2012, 9: 2016}

# Use the 'replace' method to rename values in the 'YEAR' column
df_cc_2012_2016_cleaned_yr_age['YEAR'] = df_cc_2012_2016_cleaned_yr_age['YEAR'].replace(year_mapping)
```

```python
df_cc_2012_2016_cleaned_yr_age.head()
```

```python
len(df_cc_2012_2016_cleaned_yr_age.columns)
```

```python
# Define the age bins
age_bins = [4, 10, float('inf')]  # The last bin represents 50+ age group

# Define labels for the age groups
age_labels = ['15-49', '50+']

# Use pd.cut to create age groups
df_cc_2012_2016_cleaned_yr_age['AGE_GROUP'] = pd.cut(df_cc_2012_2016_cleaned_yr_age['AGEGRP'], bins=age_bins, labels=age_labels, right=False)

# Now, df contains a new column 'AGE_GROUP' with the age group labels
#drop the old AGEGRP 
df_cc_2012_2016_cleaned_yr_age.drop(columns=['AGEGRP'], inplace=True)
```
